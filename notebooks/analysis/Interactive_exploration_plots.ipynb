# Create interactive plots

This notebook is to assist in exploring the results

```python
%matplotlib inline

import os
import sys
import warnings
import odc.geo.xr
import xarray as xr
import branca.colormap as cm
import matplotlib as mpl
from odc.geo.xr import assign_crs
```

## Analysis Parameters


```python
var = 'GPP'
version='v2.1'
base = f'/g/data/xc0/project/AusEFlux/'
```

## Open data

```python
folder = f"{base}results/AusEFlux/{version}/{var}/"
files = [f'{folder}/{i}' for i in os.listdir(folder) if i.endswith(".nc")]
files.sort()

#combine annual files into one file
ds = xr.open_mfdataset(files)
ds = assign_crs(ds, crs='EPSG:4326')
ds
```

## Plot a timestep




```python
timestep = '2025-7'
```

```python
if var=='NEE':
    cmap='Spectral_r'
    vmin, vmax= -50, 50

if var!='NEE':
    cmap='gist_earth_r'
    vmin, vmax= 0, 250

m = ds.sel(time=timestep)[var+'_median'].odc.explore(
            # tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
            # attr = 'Esri',
            # name = 'Esri Satellite',
            cmap=cmap,
            vmin=vmin,
            vmax=vmax
)

m
# cmaps = mpl.colormaps[cmap]
# colormap = cm.LinearColormap(cmaps.colors, vmin=vmin,vmax=vmax, caption=var+' gC/m2/mon')
# m.add_child(colormap)
```

```python
ds_mean = ds[var+'_median'].sel(time='2025').mean('time').compute()
```

```python
if var=='NEE':
    cmap='Spectral_r'
    vmin, vmax= -350, 350

if var!='NEE':
    cmap='gist_earth_r'
    vmin, vmax= 0, 2500

m = ds_mean.odc.explore(
            # tiles = 'https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
            # attr = 'Esri',
            # name = 'Esri Satellite',
            cmap=cmap,
            vmin=vmin,
            vmax=vmax
)

m
```

```python
import matplotlib.pyplot as plt
```

```python
ds_mean.plot.imshow(cmap=cmap,
            vmin=0,
            vmax=200, size=8);
plt.title('GPP annual mean 2025');
```
